```python
import matplotlib.pyplot as plt
import csv

def graficar_desde_csv(archivo_csv, titulo, cada_n):
    # Listas para almacenar los datos
    x = []
    y = []
    
    # Leer el archivo CSV y extraer los datos
    with open(archivo_csv, 'r') as file:
        reader = csv.reader(file)
        next(reader)  # Saltar la primera fila si contiene encabezados
        for i, row in enumerate(reader):
            if i % cada_n == 0:  # Solo añadir cada n-ésima fila
                x.append(float(row[0]))  # Convertir la primera columna a flotante y añadir a x
                y.append(float(row[1]))  # Convertir la segunda columna a flotante y añadir a y
    
    # Crear el gráfico con puntos
    plt.scatter(x, y)
    
    # Añadir título y etiquetas de los ejes
    plt.title(titulo)
    plt.xlabel('Épocas')
    plt.ylabel('Error')
    
    # Mostrar la gráfica
    plt.grid(True)
    plt.show()

# Ejemplo de uso
graficar_desde_csv('/Users/pazaramburu/Desktop/SIA-TP-S/TP3/src/multilayer_perceptron_errors.csv', 'Gráfico de Error por Época', 00)
```
